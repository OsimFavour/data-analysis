```python
# Sort the countries in ascending order and salaries in descending order
# cause we want to see the highest salaries first

df.sort_values(by=["Country", "ConvertedComp"], ascending=[True, False], inplace=True)
```

```python
df[["Country", "ConvertedComp"]].head(50)
```

```python
# maybe you are sorting results just so you can grab the
# largest or smallest values from a specific dataframe

# To see the ten highest salaries from our survey

df["ConvertedComp"] = pd.to_numeric(df["ConvertedComp"])
top_salaries = df["ConvertedComp"].nlargest(10)
```

```python
# To see the age, hours they work every week, salaries 
# paid to those with the ten highest salaries from our survey

salary = df["ConvertedComp"]
search_column = ["Age", "LanguageWorkedWith", "WorkWeekHrs", "ConvertedComp"]
top_salary_rows = df[salary.isin(top_salaries)][search_column]
```

```python
top_salary_rows.head(10)
```

```python
# another way to see the whole survey results of 
# those with the ten highest salaries from our survey

df.nlargest(10, "ConvertedComp")
```

```python
# if we wanted to see the smallest instead of the largest

df.nsmallest(1, "ConvertedComp")
```

```python
# for example: there's a question on this survey that asks
# each person what social media platform they use the most

# so if you're building an app or a website and you want to
# keep track on the most popular social media sites, you 
# might be interested on what the most popular answers to
# that question are

# to view this results, we can access the social media column
# of the website

df["SocialMedia"]
```

```python
# the social media site used the most for programmers as at
# 2018 was reddit

df["SocialMedia"].value_counts()
```

```python
# if we want to see this broken down by percentage instead of
# raw numbers, we can use:

# I used '*100', cause it gave results in percentage

df["SocialMedia"].value_counts(normalize=True) * 100
```

```python
# how can we break up the results to see the most popular social
# media site for each country

# we'll need to learn how to group our data
# the 'groupby()' operation involves some combination of 
# splitting the object, 
# applying the function and 
# combining the reults
```

```python
df["Country"].value_counts()
```

```python
# let's look at splitting the object first and group them
# by their country

country_grp = df.groupby(["Country"])
```

```python
country_grp.get_group("India")
```

```python
# Another way to do this will be by filtering

filt = df["Country"] == "United States"
df.loc[filt]
```

```python
# getting the user 'value_counts()' of users who use social media sites 
# the most in the United States

filt = df["Country"] == "United States"
# Get the social media column
df.loc[filt]["SocialMedia"].value_counts()
```

```python
# if we run this on our dataframe 'groupby' object then it
# will give us the results for all the country groups

country_grp["SocialMedia"].value_counts().head(50)
```

```python
# get the most popular used social media site for india

# this method is more useful cause now we can see this 
# result with any country without running a filter on
# each individual country in the world, so instead of
# changing the filters, we can just change the string 
# on the loc index


country_grp["SocialMedia"].value_counts().loc["Russian Federation"]
```

```python
# we can always use normalize if we want percentage results

country_grp["SocialMedia"].value_counts(normalize=True).loc["China"]
```

```python
# Getting the median of the salary column by country

country_grp["ConvertedComp"].median()
```

```python
# see the median salary for germany

country_grp["ConvertedComp"].median().loc["Germany"]
```
